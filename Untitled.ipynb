# War game Project
## A simulator of the popular card game 


```python
import random

suits=('Hearts','Diamonds ','Spades','Clubs')
rank=('Two',
            'Three',
            'Four',
            'Five',
            'Six',
            'Seven',
            'Eight',
            
            'Nine' ,
            'Ten' ,
            'Jack',
            'Queen',
            'King',
            'Ace')

card_values = {
           'Two': 2,
            'Three': 3,
            'Four': 4,
            'Five': 5,
            'Six': 6,
            'Seven': 7,
            'Eight': 8,
            'Nine': 9,
            'Ten': 10,
            'Jack': 11,
            'Queen': 12,
            'King': 13,
            'Ace': 14
        }

```

## Card class


```python
class Card:


        def __init__(self,suit,rank):
            self.suit=suit.capitalize()
            self.rank=rank.capitalize()
            self.value=card_values[self.rank]
        def __str__(self):
            return f"{self.rank} of {self.suit} and the value is {card_values[self.rank]}"
        def __int__(self):
            return self.value

    
```

## Deck Class

```python
class Deck:
    def __init__(self):
        self.all_cards=[]
        for s in suits:
            for r in rank:
                self.all_cards.append(Card(s,r))
    def shuffle(self):
        random.shuffle(self.all_cards)
    def deal_one(self):
        return self.all_cards.pop()
```

## Player Class

```python
class Player:
    def __init__(self,name):
        self.name=name
        self.all_cards=[]
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    
    def add_cards(self,new_cards):
        
        if(type(new_cards)==type([])):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
        
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
         
```

## Shuffling Deck and splitting cards among players

```python

choose=int(input("Enter the number of cards to be drawn at the time of war!!:"))
player_one=Player("One")
player_two=Player("Two")

new_deck=Deck()
new_deck.shuffle()

for x in range(26):
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one())
    

```
